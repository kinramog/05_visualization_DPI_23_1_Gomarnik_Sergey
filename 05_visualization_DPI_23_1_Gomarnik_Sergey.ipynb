Сгруппируйте все рецепты по данному признаку и для каждой группы посчитайте 2 величины: среднее количество шагов рецептов в группе и размер группы. При помощи методов структур `pandas` постройте столбчатую диаграмму, где каждый столбец означает группу (короткие, средние или длинные рецепты), а высота столбца обозначает среднее количество шагов рецептов в группе. Рядом по горизонтали разместите круговую диаграмму, на которой отображены размеры каждой из групп.

Добавьте следующие подписи:
* по горизонтальной оси под столбчатой диаграммой напишите "Группа рецептов"
* по вертикальной оси слева от столбчатой диаграммы напишите "Средняя длительность"
* над круговой диаграммой напишите "Размеры групп рецептов"

````python
easy_recipes = recipes[(recipes.minutes < 5)]
medium_recipes = recipes[(5 <= recipes.minutes) & (recipes.minutes < 50)]
hard_recipes = recipes[(recipes.minutes >= 50)]

mean_steps = [easy_recipes.n_steps.mean(), medium_recipes.n_steps.mean(), hard_recipes.n_steps.mean()]
recipe_groups_lens = [len(easy_recipes), len(medium_recipes), len(hard_recipes)]
````

````python
mean_steps_df = pd.DataFrame(mean_steps, ["легкие", "средние", "сложные"] )
recipe_groups_lens_df = pd.DataFrame(recipe_groups_lens, ["легкие", "средние", "сложные"])

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

mean_steps_df.plot(kind="bar", rot=0, ax=axes[0])
axes[0].set_xlabel('Группа рецептов')
axes[0].set_ylabel('Средняя длительность')

recipe_groups_lens_df.plot(kind='pie', ax=axes[1], y=0, title='Размеры групп рецептов', autopct='%1.1f%%')

plt.show()
````
